# Date Filtering and Time Windows

## Date Filtering

```python
# We created a function to convert the date column into a more suitable date format
def convert_date(series):
    date = pd.to_datetime(series, format='%y%m%d', errors='coerce')
    return date
```

### Account Table

```python
account['date'] = convert_date(account['date'])

account.head()
```

### Transaction Table

```python
trans['date'] = convert_date(trans['date'])

trans.head()
```

### Loan Table

```python
loan['date'] = convert_date(loan['date']) 

loan.head()
```

### Card Table

```python
# For the card table, an additional step is necesary as the initial date format differ
card.head()
```

```python
# Before applying the function, we will first only select the date part (we exclude the timestamp)
card['issued'] = card['issued'].astype(str).str[:6]
card['issued'] = convert_date(card['issued']) 

card.head()
```

## Account Features

### Lenght of Relationship in years (LOR)

### Age

Features from Age to Gender are sub-categorized (listed) under "Account Features" even though we use the df table because these features are created from columns originally from the account table before we merged them with client and disp

```python
df.info()
```

```python
# We create the Age varaible 
df.loc[:, 'age'] = 1996 - (df.loc[:, 'birth_number'] // 10000 + 1900)

df.head()
```

```python
# Quick check on the results to detect any anomaly
print(f"Age range: {df['age'].min()} to {df['age'].max()}")
print(f"Any negative ages? {(df['age'] < 0).sum()}")
```

### Age Group

```python
df.loc[:, 'age_group'] = df.loc[:, 'age'] // 10 * 10

df.head()
```

```python
#df.dtypes
```

### Birth Info

```python
df.loc[:, 'birth_year'] = df.loc[:, 'birth_number'].astype(str).str[0:2].astype(int) + 1900
df.loc[:, 'birth_month'] = df.loc[:, 'birth_number'].astype(str).str[2:4].astype(int)
df.loc[:, 'birth_day'] = df.loc[:, 'birth_number'].astype(str).str[4:6].astype(int)

# We need to correct the birthmonth by removing the gender 'mask'
df.loc[:, 'birth_month'] = df.loc[:, 'birth_month'].apply(lambda x : x-50 if x>50 else x)

df.head()
```

### Gender

```python
# Every birth number where the month is > 12 is recorded as female 'F', else male 'M'
df.loc[:, 'Gender'] = df.loc[:, 'birth_number'].apply(lambda x : 'F' if int(str(x)[2:4]) > 12 else 'M' )

df.head()
```

## District Features

```python
district.head()
```

```python
# We first start by renamming the different columns with their appropriate names 
district_features = district.rename(columns = {'A1' : 'district_id',
                                            'A2' : 'district_name',
                                            'A3' : 'region',
                                            'A4' : 'nb_inhabitants',
                                            'A5' : 'municipalities_0_499',
                                            'A6' : 'municipalities_500_1999',
                                            'A7' : 'municipalities_2000_9999',
                                            'A8' : 'municipalities_10000_plus',
                                            'A9' : 'nb_cities',
                                            'A10' : 'urban_inhabitants_ratio',
                                            'A11' : 'avg_salary',
                                            'A12' : 'unemployment_rate_95',
                                            'A13' : 'unemployment_rate_96',
                                            'A14' : 'nb_entrepreneurs_per_1000',
                                            'A15' : 'nb_crimes_95',
                                            'A16' : 'nb_crimes_96'})

# We verify if the columns were indeed renamed
district_features.columns      
```

```python
# We check if we need to cast some varaibles into a different data type
district_features.dtypes
```

```python
# We have an invalid value ('?') that should be handled right now in order to use these varaibles for feature engineering
print(district_features['unemployment_rate_95'].unique())
print(district_features['nb_crimes_95'].unique())
```

```python
# Before casting we should handle the invalid values that block the casting 
district_features['unemployment_rate_95'] = district_features['unemployment_rate_95'].replace('?', np.nan)
district_features['nb_crimes_95'] = district_features['nb_crimes_95'].replace('?', np.nan)

# We convert to numeric first
district_features['unemployment_rate_95'] = pd.to_numeric(district_features['unemployment_rate_95'], errors='coerce')
district_features['nb_crimes_95'] = pd.to_numeric(district_features['nb_crimes_95'], errors='coerce')

# We are going to use these variables for feature engineering 
# Hence can't keep them as NaN or 0 as it does not make sense (for a growth calculation for example)
# We are therefore going to replace them by the median (handles better outliers)
district_features['unemployment_rate_95'] = district_features['unemployment_rate_95'].fillna(district_features['unemployment_rate_95'].median())
district_features['nb_crimes_95'] = district_features['nb_crimes_95'].fillna(district_features['nb_crimes_95'].median())

# We check if the change was made 
district_features.dtypes
```

### Unemployment Growth

```python
# We calculate the unemployment growth from 1995 to 1996 (negative = improvement)
district_features['unemployment_growth'] = round(((district_features['unemployment_rate_96'] -  district_features['unemployment_rate_95'])
                                            / district_features['unemployment_rate_95'])*100,2)
                    
            
district_features.head()
```

### Crime Growth Rate

```python
# We calculate the crime growth rate from 1995 to 1996
district_features['crime_growth_rate'] = round(((district_features['nb_crimes_96'] -  district_features['nb_crimes_95'])
                                            / district_features['nb_crimes_95']) * 100,2)
                
    
district_features.head()
```

```python
district_features.shape
```

### Average Population per City

```python
# We calculate the average population per city in the district
district_features['population_per_city'] = round(district_features['nb_inhabitants'] / 
                                                  (district_features['nb_cities']), 2)
                
district_features.head()
```

```python
# We only select the relevant features before merging 
columns_drop = [
    'district_name',              # Text, not useful for modeling (use region instead)
    'municipalities_0_499',       # Too granular, population_per_city captures this
    'municipalities_500_1999',    
    'municipalities_2000_9999',   
    'municipalities_10000_plus',  
    'unemployment_rate_95',       # Only needed for growth calculation, not as feature
    'nb_crimes_95'                # Only needed for growth calculation, not as feature
]

district_features_to_merge = district_features.drop(columns=columns_drop)

# We merge with df 
df = pd.merge(df, district_features_to_merge, how='left', left_on='client_district_id', right_on='district_id')

df.head()
```

```python
# We verify that the shape (rows) is still intact
df.shape
```

```python
# We only select the columns with numeric values
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

numeric_cols
```

```python
# We exclude id's and targets
exclude_cols = [c for c in numeric_cols if 'target' in c or '_id' in c ]

# Filtering 
outlier_cols = [c for c in numeric_cols if c not in exclude_cols]

outlier_cols
```

```python
# We use the IQR method to detect any potential outlier 

outlier_summary = []

for col in outlier_cols:
    if df[col].std() > 0:  # Only analyze columns with variance
        Q1 = df[col].quantile(0.25) 
        Q3 = df[col].quantile(0.75) 
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Count outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        nb_outliers = len(outliers)
        
        if nb_outliers > 0:
            outlier_summary.append({
                'column': col,
                'nb_outliers': nb_outliers,
                'lower_bound': round(lower_bound, 2),
                'upper_bound': round(upper_bound, 2),
                'min': round(df[col].min(), 2),
                'max': round(df[col].max(), 2)
            })
```

```python
# We display the outlier summary
outlier_df = pd.DataFrame(outlier_summary).sort_values('nb_outliers', ascending=False)

outlier_df.head()
```
